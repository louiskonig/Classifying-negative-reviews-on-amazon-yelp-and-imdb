```python
%matplotlib inline
import numpy as np
import pandas as pd
import scipy
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns

import math
```

```python
df_imdb = pd.read_csv('imdb_labelled.txt', delimiter= '\t', header=None)
#df_imdb = pd.read_csv('yelp_labelled.txt', delimiter= '\t', header=None)
#df_imdb = pd.read_csv('amazon_cells_labelled.txt', delimiter= '\t', header=None)

df_imdb.columns = ['review', 'score']
```

```python
#tf-idf functions
def tf(word, blob):
    return blob.count(word) / len(blob)

def n_containing(word, bloblist):
    return sum(1 for blob in bloblist if word in blob)

def idf(word, bloblist):
    return math.log(len(bloblist) / (1 + n_containing(word, bloblist)))

def tfidf(word, blob, bloblist):
    return tf(word, blob) * idf(word, bloblist)
```

```python
df_imdb_0 = df_imdb[df_imdb['score'] == 0]['review']


def cleanup(title):
    word_mapping = {
        
    }
    words = title.split()
    words = [word.lower() for word in words if word.isalpha()]
    words = [word_mapping.get(word, word) for word in words]
    
    return tuple(words)
```

```python
def clean_and_compute_tfidf(dict):

    cleaned_titles = [cleanup(t) for t in dict]
    words={}
    words = {word for words in cleaned_titles for word in words}

    word_values = {word: 0.0 for word in words}

    for word in words:
        for t, cleaned in zip(dict, cleaned_titles):
            if len(cleaned) > 0:
                word_values[word] += tfidf(word, cleaned, cleaned_titles)
    return word_values
```

```python
df_imdb_tfidf = {}
df_imdb_tfidf = clean_and_compute_tfidf(df_imdb_0)
df_imdb_tfidf2 = pd.Series(df_imdb_tfidf).to_frame()
df_imdb_tfidf2.columns = ['tfidf']
df_imdb_tfidf2.sort_values(['tfidf'], ascending=[True]).to_csv('imdb_tfidf.csv')

#develop keywords based on words from above
```

```python
keywords = ['slow', 'lost', 'awful', 'lacks', 'waste', 'bor', 'unfortunate','bad','hate','waste','avoid','wors','never', \
            'awful','fails','problem','terrible','poor','stupid','hate','pathetic','negative','sloppy','flawed', \
            'terribly','predictabl','horrend','disappoint','no','n\'t', 'too']

for key in keywords:
    # Note that we add spaces around the key so that we're getting the word,
    # not just pattern matching.
    df_imdb[str(key)] = (df_imdb.review.str.contains(str(key) ,case=False)) 
```

```python
df_imdb['low_score'] = (df_imdb['score'] == 0)
```

```python
sns.heatmap(df_imdb.corr())
```

```python
data = df_imdb[keywords]
target = df_imdb['low_score']
```

```python
# Our data is binary / boolean, so we're importing the Bernoulli classifier.
from sklearn.naive_bayes import BernoulliNB
#from sklearn.naive_bayes import GaussianNB
#from sklearn.naive_bayes import MultinomialNB

# Instantiate our model and store it in a new variable.
bnb = BernoulliNB()
#72%

#bnb = MultinomialNB()
#65% for multinomial

#bnb = GaussianNB()
#63% for GaussianNB



#BernoulliNB()
#GaussianNB()
#MultinomialNB() 

# Fit our model to the data.
bnb.fit(data, target)

# Classify, storing the result in a new variable.
y_pred = bnb.predict(data)

# Display our results.
print("Number of mislabeled points out of a total {} points : {}, which is {:.0%} correct!".format(
    data.shape[0],
    (target != y_pred).sum(),
    (target == y_pred).sum()/data.shape[0]
))
```

```python
df = data

#evaluate where we got it wrong
df['y_pred'] = y_pred
df['target'] = target
df['rev'] = df_imdb['review']
df2 = df.loc[df['y_pred'] != df['target']]
```

The Bernoulli classifier above predicts with 72% - 74% accuracy whether or not a review is bad.  Feature selection involved the following process:

1) Perform TF-IDF analysis to review individual words on the set of negative reviews.

2) Manually review the output, and select a list of ~30 words/parts of words/punctuation marks that are represented.

3) Look to see if any words are highly correlated and exclude any variables from the data set.

4) Separate the data set, and the target variable (in this case a negative review)

5) Feed these 2 sets to a Bernoulli classifier.
